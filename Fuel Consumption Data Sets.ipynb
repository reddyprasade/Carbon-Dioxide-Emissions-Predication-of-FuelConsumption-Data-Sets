```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop('CO2EMISSIONS',axis=1)
X.head()
```

```python
Y= df['CO2EMISSIONS']
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
Train = pd.DataFrame(x_train,y_train)
```

```python
Train.to_csv("Train.csv")
```

```python
Test =pd.DataFrame(x_test,y_test)
```

```python
Test.to_csv('Test.csv')
```

```python
Model =LinearRegression(normalize=True)
```

```python
Model.fit(x_train,y_train)
```

```python
Model.intercept_
```

```python
Model.coef_
```

```python
Train_Score = np.round(Model.score(x_train,y_train)*100,2)
Train_Score
```

```python
Test_Score = np.round(Model.score(x_test,y_test)*100,2)
Test_Score
```

```python
New_predication = Model.predict(x_test)
```

```python
pd.DataFrame({'New_predication':New_predication,
             "ActualData":y_test,
             "Error":y_test-New_predication,
             "MSE":np.sqrt(np.mean(y_test-New_predication))})
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
r2_score(y_test,New_predication)
```

```python
mean_squared_error(y_test,New_predication)
```

```python
mean_absolute_error(y_test,New_predication)
```
